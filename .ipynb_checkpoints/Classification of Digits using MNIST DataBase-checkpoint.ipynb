```python
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
mnist = tf.keras.datasets.mnist 
(x_train,y_train) , (x_test,y_test) = mnist.load_data() 

print("shape of x_train:",x_train.shape,",y_train:",y_train.shape,",x_test:",x_test.shape,",y_test:",y_test.shape) 
```

```python
plt.figure(figsize=(3,3))
def draw(n):                          #plt.imshow: displays an array as an image
    plt.imshow(n,cmap=plt.cm.binary) #plt.cm.binary :specifies that the image should be displayed in a grayscale colormap
    plt.show()                       #renders the image on the screen
draw(x_train[1]) 
x_train[1].dtype
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Conv2D, MaxPooling2D, Flatten, Dense 
from tensorflow.keras.callbacks import ModelCheckpoint
model =Sequential() 
model.add(Conv2D(64, (5, 5), strides=2, activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())   
#To feed this data into a fully connected (dense) layer, it needs to be converted into a 1D vector

model.add(Dense(128,activation='relu')) 
model.add(Dense(128,activation='relu')) 
model.add(Dense(10,activation='softmax')) 

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

checkpoint_callback = ModelCheckpoint(
    'best_model.keras',  # filename to save the model
    monitor='val_accuracy',  # metric to monitor
    save_best_only=True,  # save only the best model
    mode='max',  # we want the maximum accuracy
    verbose=1 # print messages when saving the model
)

class_history=model.fit(x_train,y_train,epochs=10,batch_size=200, validation_split=0.10,callbacks=[checkpoint_callback]) 
```

```python
from tensorflow.keras.models import load_model
best_model = load_model('best_model.keras')
```

```python
plt.figure(figsize=(5,3))
plt.plot(class_history.history['accuracy'], label='Train Accuracy')
plt.plot(class_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Classification Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
plt.tight_layout()
```

```python
test_loss,test_acc = best_model.evaluate(x_test,y_test) 
print("loss-> ",test_loss,"\naccracy-> ",test_acc) 
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = model.predict(x_test)  #output layer which has probability of each digit
y_pred_classes = np.argmax(y_pred, axis=1)  # Get the class with the highest probability


cm = confusion_matrix(y_test, y_pred_classes)

plt.figure(figsize=(10, 8))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(10))
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for MNIST Digit Classification')
plt.show()
```

```python
predictions=model.predict([x_test]) 
print('Original label -> ',y_test[100]) 
print('predicted label -> ',np.argmax(predictions[100])) #np.argmax : returns the index of the maximum value
plt.figure(figsize=(3,3))
draw(x_test[100]) 
```
